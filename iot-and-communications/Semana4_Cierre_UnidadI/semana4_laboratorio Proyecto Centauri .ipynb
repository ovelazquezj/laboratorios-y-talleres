## 1. ¿Qué es modulación?

**Modular** significa transformar una señal digital (1s y 0s) en una forma que pueda transmitirse físicamente.

Hay muchas formas de modular:
- **ASK**: cambia la amplitud
- **FSK**: cambia la frecuencia
- **PSK**: cambia la fase

Aquí usaremos **FSK (Frequency Shift Keying)**, por ser fácil de observar y resistente al ruido.

---

## 2. Codifica un mensaje binario

Imagina que la sonda Centauri detecta este mensaje:



```python
mensaje = "Hola"
bits = ''.join(format(ord(c), '08b') for c in mensaje)
print(bits)  # 0100100001101111...
```


---

## 3. Modula con FSK

Simulemos una portadora para cada bit:



```python
import numpy as np
import matplotlib.pyplot as plt

f_0 = 1000  # frecuencia para bit 0
f_1 = 2000  # frecuencia para bit 1
bitrate = 100  # bits por segundo
sample_rate = 10000

def generate_fsk(bits):
    signal = np.array([])
    t = np.linspace(0, 1/bitrate, int(sample_rate/bitrate), endpoint=False)
    for b in bits:
        f = f_1 if b == '1' else f_0
        wave = np.sin(2*np.pi*f*t)
        signal = np.concatenate((signal, wave))
    return signal

fsk_signal = generate_fsk(bits[:8])  # solo una letra
plt.plot(fsk_signal[:500])
plt.title("Señal FSK para primeros 8 bits")
plt.show()
```


---

## 4. ¿Y si fuera una señal recibida?

Vamos a analizar su espectro:



```python
from scipy.fft import fft

N = 1000
fft_vals = np.abs(fft(fsk_signal[:N]))[:N//2]
freqs = np.fft.fftfreq(N, 1/sample_rate)[:N//2]

plt.plot(freqs, fft_vals)
plt.title("Espectro de señal FSK")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Amplitud")
plt.grid()
plt.show()
```


---

## 5. Cuantización y codificación

La cuantización es el proceso de representar una señal analógica con valores discretos. Por ejemplo, podemos tomar muestras de una señal y asignarles valores enteros.



```python
analog = np.sin(2*np.pi*440*np.linspace(0, 1, sample_rate))
cuantizada = np.round(analog * 127).astype(int)
plt.plot(cuantizada[:200])
plt.title("Ejemplo de cuantización")
plt.show()
```


Esto da lugar a **codificación digital**, el tema central de la próxima semana.
