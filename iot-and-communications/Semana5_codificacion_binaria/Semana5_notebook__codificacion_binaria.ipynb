# Notebook – Codificar en binario y enviar por HTTP (para Jupyter con Anaconda)

Este notebook servirá como acordeón para los estudiantes en la evaluación del laboratorio.

**Objetivos:**
1. Codificar un mensaje de texto en binario.
2. Enviar dicho mensaje usando HTTP POST.
3. Verificar que el mensaje llegó correctamente usando una instancia local de httpbin.

✅ Este cuaderno está diseñado para ejecutarse en una instalación local de **Jupyter Notebook (Anaconda)**.

### Paso previo
Ejecutar en terminal antes de iniciar este notebook:
```bash
docker run -d --name httpbin -p 80:80 kennethreitz/httpbin
```

````
def text_to_binary(text):
    return ' '.join(format(ord(c), '08b') for c in text)

mensaje = "Hola IoT 2025"
binario = text_to_binary(mensaje)
print("Mensaje codificado en binario:", binario)
````

````
import requests
import json

url = "http://localhost/post"
headers = {"Content-Type": "application/json"}
payload = {"mensaje_codificado": binario}

try:
    response = requests.post(url, headers=headers, data=json.dumps(payload))
    if response.status_code == 200:
        respuesta = response.json()
        print("\n✔️ Respuesta recibida del servidor:")
        print(json.dumps(respuesta["json"], indent=4))
    else:
        print("❌ Error en la solicitud:", response.status_code)
except requests.exceptions.ConnectionError:
    print("🚫 No se pudo conectar a http://localhost/post. Asegúrate de tener httpbin corriendo con Docker.")
````
