## Ejercicio 1 – Arreglo Unidimensional y Merge Sort
**Problema:** Formar una fila de abordaje para un autobús con base en el número de ticket.

**Descripción:** Cada pasajero recibe un ticket con un número aleatorio del 1 al 40. El objetivo es organizar a los pasajeros en orden creciente de tickets utilizando el algoritmo Merge Sort.

```
# Algoritmo de Merge Sort
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left = arr[:mid]
        right = arr[mid:]

        merge_sort(left)
        merge_sort(right)

        i = j = k = 0

        while i < len(left) and j < len(right):
            if left[i] < right[j]:
                arr[k] = left[i]
                i += 1
            else:
                arr[k] = right[j]
                j += 1
            k += 1

        while i < len(left):
            arr[k] = left[i]
            i += 1
            k += 1

        while j < len(right):
            arr[k] = right[j]
            j += 1
            k += 1
```

```
# Paso 1: Generar tickets aleatorios
import random

tickets_pasajeros = random.sample(range(1, 41), 10)
print("Tickets desordenados:", tickets_pasajeros)

# Paso 2: Ordenar
merge_sort(tickets_pasajeros)

# Paso 3: Visualización ordenada
def mostrar_fila(tickets):
    print("\nFila ordenada para abordar el autobús (por número de ticket):")
    print("="*45)
    for i, ticket in enumerate(tickets, start=1):
        print(f"Pasajero {i:2d} → Ticket #{ticket:02d}")
    print("="*45)

mostrar_fila(tickets_pasajeros)
```

## Ejercicio 3 – Pila FILO y búsqueda binaria
**Problema:** Simular pila de procesos del sistema operativo y aplicar búsqueda binaria por PID.

```
# Generar 150 PIDs únicos
pids = random.sample(range(1000, 5000), 150)
stack_procesos = []

print("\nCargando procesos en pila:")
for pid in pids:
    stack_procesos.append(pid)
    print(f"Push → PID {pid}")

# Búsqueda binaria
def busqueda_binaria(arr, objetivo):
    inicio = 0
    fin = len(arr) - 1
    while inicio <= fin:
        medio = (inicio + fin) // 2
        if arr[medio] == objetivo:
            return medio
        elif arr[medio] < objetivo:
            inicio = medio + 1
        else:
            fin = medio - 1
    return -1

# Prueba de búsqueda
pids_ordenados = sorted(pids)
target_pid = pids[75]
print(f"\nBuscando PID {target_pid}...")
resultado = busqueda_binaria(pids_ordenados, target_pid)
if resultado != -1:
    print(f"→ PID {target_pid} encontrado en posición {resultado}.")
else:
    print("→ PID no encontrado.")
```
